```python
#Importing necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading in the data
data_2015 = pd.read_csv("2015.csv")
data_2016 = pd.read_csv("2016.csv")
data_2017 = pd.read_csv("2017.csv")
data_2018 = pd.read_csv("2018.csv")
data_2019 = pd.read_csv("2019.csv")
data_2020 = pd.read_csv("2020.csv")
data_2021 = pd.read_csv("2021.csv")
data_2022 = pd.read_csv("2022.csv")
```

```python
#fixing 2015 data

#dropping columns
columns_to_drop = ['Region', 'Standard Error', 'Dystopia Residual', 'Family']
data_2015 = data_2015.drop(columns = columns_to_drop)

#renaming columns
col_names = {
    'Happiness Rank': 'Rank',
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP Per Capita',
    'Health (Life Expectancy)': 'Life Expectancy',
    'Trust (Government Corruption)': 'Government Corruption'
}
data_2015 = data_2015.rename(columns=col_names)

#re-ordering columns
order = ['Rank', 'Country', 'Score', 'GDP Per Capita', 'Life Expectancy','Freedom', 'Generosity', 'Government Corruption']
data_2015 = data_2015[order]

data_2015
```

```python
#fixing 2016 data

#dropping columns
columns_to_drop = ['Region', 'Lower Confidence Interval', 'Dystopia Residual', 'Family', 'Upper Confidence Interval']
data_2016 = data_2016.drop(columns = columns_to_drop)

#renaming columns
col_names = {
    'Happiness Rank': 'Rank',
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP Per Capita',
    'Health (Life Expectancy)': 'Life Expectancy',
    'Trust (Government Corruption)': 'Government Corruption'
}

data_2016 = data_2016.rename(columns=col_names)

#re-ordering columns
order = ['Rank', 'Country', 'Score', 'GDP Per Capita', 'Life Expectancy','Freedom', 'Generosity', 'Government Corruption']
data_2016 = data_2016[order]

data_2016
```

```python
#fixing 2017 data issues

#dropping columns
columns_to_drop = ['Whisker.high', 'Whisker.low', 'Dystopia.Residual', 'Family']
data_2017 = data_2017.drop(columns = columns_to_drop)

#renaming columns to common format
col_names = {
    'Country': 'Country',
    'Happiness.Rank': 'Rank',
    'Happiness.Score': 'Score',
    'Economy..GDP.per.Capita.': 'GDP Per Capita',
    'Health..Life.Expectancy.': 'Life Expectancy',
    'Freedom': 'Freedom',
    'Generosity': 'Generosity',
    'Trust..Government.Corruption.': 'Government Corruption'
}

data_2017 = data_2017.rename(columns=col_names)
#re-ordering columns
order = ['Rank', 'Country', 'Score', 'GDP Per Capita', 'Life Expectancy','Freedom', 'Generosity', 'Government Corruption']
data_2017 = data_2017[order]
data_2017
```

```python
#checking 2018 columns

#renaming columns to common format
col_names = {
    'Country or region': 'Country',
    'Overall rank': 'Rank',
    'GDP per capita': 'GDP Per Capita',
    'Healthy life expectancy': 'Life Expectancy',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Government Corruption'
}

data_2018 = data_2018.rename(columns=col_names)

#dropping columns
data_2018 = data_2018.drop(columns = 'Social support')
data_2018
```

```python
#Fixing 2019 data

#dropping columns
data_2019 = data_2019.drop(columns = 'Social support')

#renaming columns
col_names = {
    'Overall rank': 'Rank',
    'Country or region': 'Country',
    'Happiness.Score': 'Score',
    'GDP per capita': 'GDP Per Capita',
    'Healthy life expectancy': 'Life Expectancy',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Government Corruption'
}
data_2019 = data_2019.rename(columns=col_names)

data_2019
```

```python
#fixing 2020 columns -- DONEEEE

#dropping columns
columns_to_drop = ['Regional indicator', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Logged GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Dystopia + residual', 'Social support', 'Ladder score in Dystopia', 'Explained by: Social support']
data_2020 = data_2020.drop(columns = columns_to_drop)

#renaming columns
col_names = {
    'Ladder score' : 'Score',
    'Country name' : 'Country',
    'Overall rank' : 'Rank',
    'Explained by: Log GDP per capita': 'GDP Per Capita',
    'Explained by: Healthy life expectancy': 'Life Expectancy',
    'Explained by: Freedom to make life choices': 'Freedom',
    'Explained by: Generosity': 'Generosity',
    'Explained by: Perceptions of corruption': 'Government Corruption'
}

data_2020 = data_2020.rename(columns=col_names)

#Adding rank column
data_2020['Rank'] = data_2020.index + 1

#re-ordering columns
order = ['Rank', 'Country', 'Score', 'GDP Per Capita', 'Life Expectancy','Freedom', 'Generosity', 'Government Corruption']
data_2020 = data_2020[order]

data_2020
```

```python
#fixing 2021 data

#saving region col
data_2021_region = data_2021

#dropping columns
columns_to_drop = ['Regional indicator', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Logged GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Dystopia + residual', 'Social support', 'Ladder score in Dystopia', 'Explained by: Social support']
data_2021 = data_2021.drop(columns = columns_to_drop)

#renaming columns
col_names = {
    'Ladder score' : 'Score',
    'Country name' : 'Country',
    'Overall rank' : 'Rank',
    'Explained by: Log GDP per capita': 'GDP Per Capita',
    'Explained by: Healthy life expectancy': 'Life Expectancy',
    'Explained by: Freedom to make life choices': 'Freedom',
    'Explained by: Generosity': 'Generosity',
    'Explained by: Perceptions of corruption': 'Government Corruption'
}

data_2021 = data_2021.rename(columns=col_names)

#Adding rank column
data_2021['Rank'] = data_2021.index + 1

#re-ordering columns
order = ['Rank', 'Country', 'Score', 'GDP Per Capita', 'Life Expectancy','Freedom', 'Generosity', 'Government Corruption']
data_2021 = data_2021[order]

data_2021
```

```python
#Fixing 2022 data

#dropping columns
columns_to_drop = ['Whisker-high', 'Whisker-low', 'Dystopia (1.83) + residual', 'Explained by: Social support']
data_2022 = data_2022.drop(columns = columns_to_drop)

#renaming columns
col_names = {
    'RANK' : 'Rank',
    'Happiness score': 'Score',
    'Explained by: GDP per capita': 'GDP Per Capita',
    'Explained by: Healthy life expectancy': 'Life Expectancy',
    'Explained by: Freedom to make life choices': 'Freedom',
    'Explained by: Generosity': 'Generosity',
    'Explained by: Perceptions of corruption': 'Government Corruption'
}

data_2022 = data_2022.rename(columns=col_names)

#dropping Na's
data_2022 = data_2022.dropna()

# Remove asterisks from the 'Country' column
data_2022['Country'] = data_2022['Country'].str.replace('*', '')

#changing numbers to american format
numeric_columns = ['Score', 'GDP Per Capita', 'Life Expectancy', 'Freedom', 'Generosity', 'Government Corruption']
for col in numeric_columns:
    data_2022[col] = data_2022[col].str.replace(',', '.').astype(float)


data_2022
```

```python
# Add 'Year' column to each dataframe
data_2015['Year'] = 2015
data_2016['Year'] = 2016
data_2017['Year'] = 2017
data_2018['Year'] = 2018
data_2019['Year'] = 2019
data_2020['Year'] = 2020
data_2021['Year'] = 2021
data_2022['Year'] = 2022


#Joining the data sets together

dataframes = [data_2015, data_2016, data_2017, data_2018, data_2019, data_2020, data_2021, data_2022]

# Concatenate the dataframes along the rows
happiness = pd.concat(dataframes, ignore_index=True)

# Display the concatenated dataframe
happiness    #Now the ranking is off. Is this an issue??


```

```python
# US scores
happiness[happiness['Country']=='United States']
```

```python
#Bar plot to show top 10 happiest countries in 2022
top_n = 10
plt.figure(figsize=(12, 6))
sns.barplot(x='Score', y='Country', data=happiness[happiness['Year'] == 2022].nlargest(top_n, 'Score'), palette='viridis')
plt.title(f'Top {top_n} Countries by Happiness Score in 2022')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('top_countries_2022.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
#Bar plot to show top 10 happiest countries in 2022
bottom_n = 10
plt.figure(figsize=(12, 6))
sns.barplot(x='Score', y='Country', data=happiness[happiness['Year'] == 2022].nsmallest(bottom_n, 'Score'), palette='viridis')
plt.title(f'Bottom {bottom_n} Countries by Happiness Score in 2022')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.savefig('bottom_countries_2022.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
# Region comparison plot 
df = data_2021_region.sort_values(by='Ladder score')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ladder score', y='Regional indicator', data=df) 
plt.title('Happiness Score Comparison by Region in 2021')
plt.xlabel("Happiness Score")  
plt.ylabel("Region")
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('region_comparison_2021.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
#Creating histogram for happiness scores in 2022
happiness_scores_2022 = data_2022['Score']

#create the bins
score_ranges = [2, 3, 4, 5, 6, 7, 8]

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(happiness_scores_2022, bins=score_ranges, edgecolor='black', color='skyblue')
plt.title('Distribution of Happiness Scores in 2022')
plt.xlabel('Happiness Score Ranges')
plt.ylabel('Number of Countries')
plt.xticks(score_ranges)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('happiness_scores_distribution_2022.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
# Line plot to show the trend in average happiness 
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Score', data=happiness, errorbar = None)
plt.title('Average Happiness Score Over Time')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('average_happiness_score_over_time.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
#Factors we want
factors = ['Score', 'GDP Per Capita', 'Life Expectancy', 'Freedom', 'Generosity', 'Government Corruption']

# Calculate correlation matrix
correlation_matrix = happiness[factors].corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix: Factors and Happiness Score')

#saving image
plt.savefig('correlation_matrix_factors_happiness_score.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
# Identify the top 3 ranking countries for the year 2022
top_countries_2022 = happiness[happiness['Year'] == 2022].nlargest(3, 'Score')

# Transpose the DataFrame for easier plotting
transposed_data = top_countries_2022.set_index('Country')[['GDP Per Capita', 'Life Expectancy', 'Freedom', 'Generosity', 'Government Corruption']].T

# Create a grouped bar chart 
plt.figure(figsize=(12, 8))
transposed_data.plot(kind='bar', colormap='viridis', position=1, width=0.5) 
plt.title('Top 3 Countries\' Factor Scores in 2022')
plt.xlabel('Factor')
plt.ylabel('Factor Score')
plt.legend(title='Country')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('Top_countries_2022_factors.svg', format='svg', bbox_inches='tight')
plt.show()
```

```python
# Identify the bottom 3 ranking countries for the year 2022
bottom_countries_2022 = happiness[happiness['Year'] == 2022].nsmallest(3, 'Score')

# Transpose the DataFrame for easier plotting
transposed_data_bottom = bottom_countries_2022.set_index('Country')[['GDP Per Capita', 'Life Expectancy', 'Freedom', 'Generosity', 'Government Corruption']].T

# Create a grouped bar chart 
plt.figure(figsize=(12, 8))
transposed_data_bottom.plot(kind='bar', colormap='viridis', position=1, width=0.5)  
plt.title('Bottom 3 Countries\' Factor Scores in 2022')
plt.xlabel('Factor')
plt.ylabel('Factor Score')
plt.legend(title='Country')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

#saving image
plt.savefig('bottom_countries_2022_factors.svg', format='svg', bbox_inches='tight')
plt.show()
```
